```python
train_data = pd.read_csv('/kaggle/input/titanic/train.csv')
test_data = pd.read_csv('/kaggle/input/titanic/test.csv')
train_data.head(20)
print('\n Obs. in training: ', train_data.shape[0], '\n Obs. in test: ', test_data.shape[0], '\n Features: ', train_data.shape[1]-2 )
```

```python
train_data = train_data.drop(["PassengerId","Name","Embarked"],axis=1)
```

```python
train_data
```

```python
train_ds = tfdf.keras.pd_dataframe_to_tf_dataset(train_data, label="Survived")
test_ds = tfdf.keras.pd_dataframe_to_tf_dataset(test_data)
model = tfdf.keras.RandomForestModel(
           num_trees=500,
           categorical_algorithm = 'CART',
           min_examples=10,
           winner_take_all=True,
           max_depth=5
           )
    
model.compile(metrics=["accuracy"])
model.fit(train_ds)
```

```python
#预测
predictions =  model.predict(test_ds)
predictions_n = (predictions > 0.5).astype('int')
predictions_n
```

```python
output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': predictions_n.squeeze()})
output.to_csv('submission.csv', index=False)
```
